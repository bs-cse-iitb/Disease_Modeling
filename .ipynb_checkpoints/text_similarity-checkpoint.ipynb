```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('distiluse-base-multilingual-cased-v2')

def CheckSimilarity(doc1,doc2):
  # Compute the embeddings for the text in the documents
  embeddings1 = model.encode(doc1, convert_to_tensor=True)
  embeddings2 = model.encode(doc2, convert_to_tensor=True)

  # Compute the cosine similarity between the embeddings
  cosine_scores = util.pytorch_cos_sim(embeddings1, embeddings2)

  return cosine_scores.item()
```

```python
# Open the file for reading
KalaAzharUdaan = ''
with open('KalaAzharUdaan.txt', 'r') as file:
  for line in file:
    KalaAzharUdaan += line.strip()

KalaAzharIDSP = ''
with open('KalaAzharIDSP.txt', 'r') as file:
  for line in file:
    KalaAzharIDSP += line.strip()

FilariasisUdaan = ''
with open('FilariasisUdaan.txt', 'r') as file:
  for line in file:
    FilariasisUdaan += line.strip()

FilariasisIDSP = ''
with open('FilariasisIDSP.txt', 'r') as file:
  for line in file:
    FilariasisIDSP += line.strip()
  
```
